```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() 
```

```python
netflix = pd.read_csv('cleaned_data.csv')
netflix.head()
```

```python

basiccost = pd.DataFrame(netflix["Cost Per Month - Basic ($)"])
standardcost = pd.DataFrame(netflix["Cost Per Month - Standard ($)"])
premiumcost = pd.DataFrame(netflix["Cost Per Month - Premium ($)"])
```

```python
density = pd.DataFrame(netflix["Density (P/Km²)"])
```

```python
jointDF = pd.concat([basiccost, density], axis = 1).reindex(basiccost.index)
jointDF
```

```python
sb.jointplot(data = jointDF, x = "Cost Per Month - Basic ($)", y = "Density (P/Km²)", height = 10)
jointDF.corr()
```

```python
jointDE = pd.concat([standardcost, density], axis = 1).reindex(standardcost.index)
jointDE
```

```python
sb.jointplot(data = jointDE, x = "Cost Per Month - Standard ($)", y = "Density (P/Km²)", height = 10)
jointDE.corr()
```

```python
jointDC = pd.concat([premiumcost, density], axis = 1).reindex(premiumcost.index)
jointDC
```

```python
sb.jointplot(data = jointDC, x = "Cost Per Month - Premium ($)", y = "Density (P/Km²)", height = 10)
jointDC.corr()
```
