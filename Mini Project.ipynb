```python
training = data.sample(frac = 0.8, random_state=180)
test = data.drop(training.index)
print("Training Set: ", training.shape)
print("Test Set: ", test.shape)
```

```python
basicTrain = pd.DataFrame(training["Cost Per Month - Basic ($)"])
standardTrain = pd.DataFrame(training["Cost Per Month - Standard ($)"])
premiumTrain = pd.DataFrame(training["Cost Per Month - Premium ($)"])
gdpTrain = pd.DataFrame(training["2018"])
popDensityTrain = pd.DataFrame(training["Density (P/Km²)"])

basicTest = pd.DataFrame(test["Cost Per Month - Basic ($)"])
standardTest = pd.DataFrame(test["Cost Per Month - Standard ($)"])
premiumTest = pd.DataFrame(test["Cost Per Month - Premium ($)"])
gdpTest = pd.DataFrame(test["2018"])
popDensityTest = pd.DataFrame(test["Density (P/Km²)"])
```

```python
# Import LinearRegression model from Scikit-Learn
from sklearn.linear_model import LinearRegression

# Create a Linear Regression object
linreg1 = LinearRegression()
linreg2 = LinearRegression()
linreg3 = LinearRegression()

linreg1.fit(gdpTrain, basicTrain)
linreg2.fit(gdpTrain, standardTrain)
linreg3.fit(gdpTrain, premiumTrain)

print('Intercept \t: b = ', linreg1.intercept_, linreg2.intercept_, linreg3.intercept_,)
print('Coefficients \t: a = ', linreg1.coef_, linreg2.coef_, linreg3.coef_,)
```

```python
regline_x = gdpTrain
regline_y = linreg1.intercept_ + linreg1.coef_ * gdpTrain

f = plt.figure(figsize=(16, 8))
plt.scatter(gdpTrain, basicTrain)
plt.plot(regline_x, regline_y, 'r-', linewidth = 3)
plt.show()
from sklearn.metrics import accuracy_score

print("Explained Variance (R^2) \t:", linreg1.score(gdpTrain, basicTrain))
```

```python
regline_x = gdpTrain
regline_y = linreg2.intercept_ + linreg2.coef_ * gdpTrain

f = plt.figure(figsize=(16, 8))
plt.scatter(gdpTrain, standardTrain)
plt.plot(regline_x, regline_y, 'r-', linewidth = 3)
plt.show()

print("Explained Variance (R^2) \t:", linreg2.score(gdpTrain, standardTrain))
```

```python
regline_x = gdpTrain
regline_y = linreg3.intercept_ + linreg3.coef_ * gdpTrain

f = plt.figure(figsize=(16, 8))
plt.scatter(gdpTrain, premiumTrain)
plt.plot(regline_x, regline_y, 'r-', linewidth = 3)
plt.show()

print("Explained Variance (R^2) \t:", linreg3.score(gdpTrain, premiumTrain))
```

```python
from sklearn.metrics import accuracy_score

basicPred = linreg1.predict(gdpTrain)
standardPred = linreg2.predict(gdpTrain)
premiumPred = linreg3.predict(gdpTrain)

basictestPred = linreg1.predict(gdpTest)
standardtestPred = linreg2.predict(gdpTest)
premiumtestPred = linreg3.predict(gdpTest)

def mean_sq_err(actual, predicted):
    return np.mean(np.square(np.array(actual) - np.array(predicted)))

mse1 = mean_sq_err(basicTrain, basicPred)
mse2 = mean_sq_err(standardTrain, standardPred)
mse3 = mean_sq_err(premiumTrain, premiumPred)

mse4 = mean_sq_err(basicTest, basictestPred)
mse5 = mean_sq_err(standardTest, standardtestPred)
mse6 = mean_sq_err(premiumTest, premiumtestPred)
print("Mean Squared Error (Train) (MSE) \t:", mse1, mse2, mse3)
print("Mean Squared Error (Test) (MSE) \t:", mse4, mse5, mse6)
print("Root Mean Squared Error (Train) (RMSE) \t:)", np.sqrt(mse1), np.sqrt(mse2), np.sqrt(mse3))
print("Root Mean Squared Error (Test) (RMSE) \t:)", np.sqrt(mse4), np.sqrt(mse5), np.sqrt(mse6))
```

```python
# Create a Linear Regression object
poplinreg1 = LinearRegression()
poplinreg2 = LinearRegression()
poplinreg3 = LinearRegression()

poplinreg1.fit(popDensityTrain, basicTrain)
poplinreg2.fit(popDensityTrain, standardTrain)
poplinreg3.fit(popDensityTrain, premiumTrain)

print('Intercept \t: b = ', poplinreg1.intercept_, poplinreg2.intercept_, poplinreg3.intercept_,)
print('Coefficients \t: a = ', poplinreg1.coef_, poplinreg2.coef_, poplinreg3.coef_,)
```

```python
regline_x = popDensityTrain
regline_y = poplinreg3.intercept_ + poplinreg3.coef_ * popDensityTrain

f = plt.figure(figsize=(16, 8))
plt.scatter(popDensityTrain, premiumTrain)
plt.plot(regline_x, regline_y, 'r-', linewidth = 3)
plt.show()

print("Explained Variance (R^2) \t:", poplinreg3.score(popDensityTrain, premiumTrain))
```

```python
basicPred2 = poplinreg1.predict(popDensityTrain)
standardPred2 = poplinreg2.predict(popDensityTrain)
premiumPred2 = poplinreg3.predict(popDensityTrain)

basictestPred2 = poplinreg1.predict(popDensityTrain)
standardtestPred2 = poplinreg2.predict(popDensityTrain)
premiumtestPred2 = poplinreg3.predict(popDensityTrain)

mserr1 = mean_sq_err(basicTrain, basicPred2)
mserr2 = mean_sq_err(standardTrain, standardPred2)
mserr3 = mean_sq_err(premiumTrain, premiumPred2)

mserr4 = mean_sq_err(basicTest, basictestPred2)
mserr5 = mean_sq_err(standardTest, standardtestPred2)
mserr6 = mean_sq_err(premiumTest, premiumtestPred2)

print("Mean Squared Error (MSE) \t:", mserr1, mserr2, mserr3, mserr4, mserr5, mserr6)
print("Root Mean Squared Error (RMSE) \t:)", np.sqrt(mserr1), np.sqrt(mserr2), np.sqrt(mserr3), np.sqrt(mserr4), np.sqrt(mserr5), np.sqrt(mserr6))
```
