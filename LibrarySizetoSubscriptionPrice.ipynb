```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set()
```

```python
netflixdata = pd.read_csv('cleaned_data.csv')
netflixdata.head()
```

```python
library = pd.DataFrame(netflixdata['Total Library Size'])
```

```python
f = plt.figure(figsize=(24, 4))
sb.boxplot(data = library, orient = "h")
```

```python
numDF = pd.DataFrame(netflixdata[["Cost Per Month - Basic ($)", "Cost Per Month - Standard ($)", "Cost Per Month - Premium ($)"]])
basiccost = pd.DataFrame(netflixdata["Cost Per Month - Basic ($)"])
standardcost = pd.DataFrame(netflixdata["Cost Per Month - Standard ($)"])
premiumcost = pd.DataFrame(netflixdata["Cost Per Month - Premium ($)"])
```

```python
f = plt.figure(figsize=(16, 8))
sb.boxplot(data = numDF, orient = "h")
```

```python
jointDF = pd.concat([basiccost, library], axis = 1).reindex(basiccost.index)
jointDF
sb.jointplot(data = jointDF, x = "Cost Per Month - Basic ($)", y = "Total Library Size", height = 12)
jointDF.corr()
```

```python
jointDF = pd.concat([standardcost, library], axis = 1).reindex(standardcost.index)
jointDF
sb.jointplot(data = jointDF, x = "Cost Per Month - Standard ($)", y = "Total Library Size", height = 12)
jointDF.corr()
```

```python
jointDF = pd.concat([premiumcost, library], axis = 1).reindex(premiumcost.index)
jointDF
sb.jointplot(data = jointDF, x = "Cost Per Month - Premium ($)", y = "Total Library Size", height = 12)
jointDF.corr()
```
